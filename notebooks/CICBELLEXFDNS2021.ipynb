# 2. Data preprocessing

1. Split dataset to four subsets:
    * Heavy stateful
    * Heavy stateless
    * Light stateful
    * Light stateless
2. Do a train/test split of the subsets.
3. Use TargetEncoder to encode categorical columns.
4. Normalize data using [StandardScaler](https://www.geeksforgeeks.org/machine-learning/what-is-standardization-in-machine-learning/)

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, TargetEncoder
```

```python
#heavy_stateful.drop(['unique_ttl'], axis = 1, inplace = True)
heavy_stateless.drop(['timestamp'], axis = 1, inplace = True)
#light_stateful.drop(['unique_ttl'], axis = 1, inplace = True)
light_stateless.drop(['timestamp'], axis = 1, inplace = True)
```

```python
# Shuffle all rows
heavy_stateful = heavy_stateful.sample(frac=1).reset_index(drop=True)
heavy_stateless = heavy_stateless.sample(frac=1).reset_index(drop=True)
light_stateful = light_stateful.sample(frac=1).reset_index(drop=True)
light_stateless = light_stateless.sample(frac=1).reset_index(drop=True)
```

```python
le = LabelEncoder()

X_heavy_stateful = heavy_stateful.drop(['Class'], axis=1)
y_heavy_stateful = heavy_stateful['Class']
y_heavy_stateful = le.fit_transform(y_heavy_stateful)

stateful_categorical_cols = X_heavy_stateful.select_dtypes(include='object').columns
stateful_numerical_cols = X_heavy_stateful.select_dtypes(exclude='object').columns

X_heavy_stateful_train, X_heavy_stateful_test, y_heavy_stateful_train, y_heavy_stateful_test = train_test_split(X_heavy_stateful, y_heavy_stateful, test_size=0.2, random_state=42)
```

```python
%%time

encoder = TargetEncoder(
    target_type='multiclass',
    cv=5,
    smooth='auto',
    random_state=42
)

X_heavy_stateful_train_encoded = encoder.fit_transform(X_heavy_stateful_train[stateful_categorical_cols], y_heavy_stateful_train)
X_heavy_stateful_test_encoded = encoder.transform(X_heavy_stateful_test[stateful_categorical_cols])
```

```python
X_heavy_stateful_train_encoded = pd.DataFrame(X_heavy_stateful_train_encoded)
X_heavy_stateful_test_encoded = pd.DataFrame(X_heavy_stateful_test_encoded)
```

```python
X_heavy_stateful_train_final = pd.concat([
    X_heavy_stateful_train_encoded.reset_index(drop=True),
    X_heavy_stateful_train[stateful_numerical_cols].reset_index(drop=True)
], axis = 1)

X_heavy_stateful_test_final = pd.concat([
    X_heavy_stateful_test_encoded.reset_index(drop=True),
    X_heavy_stateful_test[stateful_numerical_cols].reset_index(drop=True)
], axis = 1)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
%%time

scaler = StandardScaler()

X_heavy_stateful_train_final.columns = X_heavy_stateful_train_final.columns.astype(str)
X_heavy_stateful_test_final.columns = X_heavy_stateful_test_final.columns.astype(str)

X_heavy_stateful_train_final = scaler.fit_transform(X_heavy_stateful_train_final)
X_heavy_stateful_test_final = scaler.fit_transform(X_heavy_stateful_test_final)
```

```python
%%time

params = {
    'max_depth': 9,
    'min_child_weight': 3,
    'learning_rate': 0.1,
    'n_estimators': 5000,
    'subsample': 0.8,
    'colsample_bytree': 0.8,
    'gamma': 0.1,
    'objective': 'binary:logistic',
    'eval_metric': 'logloss',
    'random_state': 42
}

model = xgb.XGBClassifier(**params)
model.fit(X_heavy_stateful_train_final, y_heavy_stateful_train)
```

```python
y_heavy_stateful_pred = model.predict(X_heavy_stateful_test_final)
```

```python
from sklearn.metrics import accuracy_score, root_mean_squared_error, mean_squared_error, mean_absolute_error, precision_score, confusion_matrix, multilabel_confusion_matrix, classification_report, f1_score, precision_score, recall_score
```

```python
results_dict = {}
```

```python
accuracy = accuracy_score(y_true=y_heavy_stateful_test, y_pred=y_heavy_stateful_pred)
precision = precision_score(y_true=y_heavy_stateful_test, y_pred=y_heavy_stateful_pred, average="weighted")
f1 = f1_score(y_true=y_heavy_stateful_test, y_pred=y_heavy_stateful_pred, average="weighted")
recall = recall_score(y_true=y_heavy_stateful_test, y_pred=y_heavy_stateful_pred, average="weighted")
mae = mean_absolute_error(y_true=y_heavy_stateful_test,y_pred=y_heavy_stateful_pred)
mse = mean_squared_error(y_true=y_heavy_stateful_test,y_pred=y_heavy_stateful_pred) 
rmse = root_mean_squared_error(y_true=y_heavy_stateful_test,y_pred=y_heavy_stateful_pred)
```

```python
results = [accuracy, precision, f1, recall, mae, mse, rmse]
results_dict["XGB_heavy_stateful_3"] = results
```
